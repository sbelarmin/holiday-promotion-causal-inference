# Causal Evaluation Summary: 40% Promotion

## Objective

Estimate the incremental margin impact of a 40% promotional campaign using observational data where treatment was targeted (not randomized).

The primary outcome is **margin dollars**.
The secondary outcome is revenue.

---

# Causal Ladder Summary

We estimated the treatment effect using progressively stronger identification strategies.

---

## 1. Naïve Post-Period Comparison

**Method:** Compare exposed vs. non-exposed users during the promo window.

**Result:** -5.61

**Issue:** Biased due to selection.  
Exposed users had higher prior spend and engagement before the promo.

---

## 2. Difference-in-Differences (DiD)

**Method:** Compare pre-to-post changes for treated vs control.

**Result:** -4.60

**Interpretation:** Adjusts for time-invariant differences between groups.

This removes baseline selection bias but assumes parallel trends.

---

## 3. Two-Way Fixed Effects (TWFE)

**Method:** DiD with user and date fixed effects.

**Result:** -4.60

**Interpretation:** Controls for:
- Time-invariant user heterogeneity
- Daily macro shocks

Estimate unchanged from basic DiD.

---

## 4. Propensity Score Weighting (IPW)

**Goal:** Rebalance treated and control users on observable characteristics.

**Finding:**
- Propensity scores clustered near 1.0
- Severe overlap failure
- Covariate balance did not improve after weighting

**Conclusion:** IPW not credible due to lack of common support.

---

## 5. Weighted DiD

**Method:** DiD with inverse probability weights.

**Result:** -4.59

**Interpretation:** Nearly identical to unweighted DiD.

Weighting did not materially alter the estimate because overlap was poor.

---

## 6. Event Study (Two-Way FE)

**Purpose:** Test parallel trends and examine dynamic effects.

**Findings:**
- Pre-period coefficients are small and statistically insignificant.
- No strong pre-trend violation detected.
- Margin drops sharply at promo start (evt_0 ≈ -7.9).
- Post-period effects remain consistently negative (-5 to -6 range).

**Conclusion:** Parallel trends assumption appears broadly reasonable.

---

# Comparison to Ground Truth (Synthetic Data)

| Estimator | Estimate |
|-----------|----------|
| True ATT | -6.54 |
| Naïve | -5.61 |
| DiD | -4.60 |
| TWFE | -4.60 |
| Weighted DiD | -4.59 |

Remaining bias relative to ATT likely reflects:
- Treatment effect heterogeneity
- Time-varying structural differences
- Imperfect identification in observational setting

---

# Business Interpretation

The 40% promotion:

- Increased unit sales
- Reduced margin significantly
- Produced an incremental margin impact of approximately **-$4.6 per exposed user-day**

Results are robust across multiple identification strategies.

Parallel trends diagnostics support the credibility of the DiD framework.

---

# Key Takeaways

1. Selection bias exists when marketing targets high-value users.
2. DiD meaningfully reduces bias relative to naïve estimates.
3. IPW fails when overlap assumptions are violated.
4. Event study supports the causal interpretation of DiD.
5. Strong targeting structures limit the effectiveness of reweighting approaches.

---

# Final Assessment

Under standard DiD assumptions, the promotion appears economically destructive in the short term.

Long-term value (retention, LTV, acquisition) would be required to justify the observed margin loss.

# Executive Summary  
## Causal Evaluation of 40% Promotional Campaign

### Objective

Assess whether a 40% discount campaign generated incremental margin dollars.

Primary outcome: **Margin dollars**  
Secondary outcome: Revenue  

The campaign was observational and targeted — not randomized.

---

## Key Finding

After adjusting for customer mix and seasonal effects, the promotion reduced margin by approximately:

### **–$4.6 per exposed user-day**

At campaign scale, this represents a material short-term margin loss.

---

## What Happened

- Units increased during the promotion.
- Revenue increased modestly.
- Margin rate declined sharply.
- Net incremental margin impact was negative.

The discount depth overwhelmed incremental volume gains.

---

## Credibility of Results

We tested the effect using multiple methods:

- Naïve comparison
- Difference-in-Differences
- Two-Way Fixed Effects
- Propensity Score Weighting
- Weighted DiD
- Event Study diagnostics

Results were consistent across specifications.

Parallel trends diagnostics show no major pre-period divergence.

---

## Important Caveat

The campaign was highly targeted toward higher-value customers.  
Statistical reweighting methods (IPW) were limited due to lack of overlap.

However, dynamic event study analysis supports the DiD estimate.

---

## Business Implication

The promotion appears economically destructive in the short term.

To justify the strategy, long-term benefits would need to exceed:

- Short-term margin losses
- Discount-driven profitability erosion

Without evidence of sustained retention or LTV gains, the discount depth should be reconsidered.

---

## Recommendation

Evaluate:

- Lower discount depth (e.g., 20–25%)
- More targeted margin-protected segmentation
- Long-term customer retention impact

Short-term economics alone do not support the 40% strategy.

# Methods Appendix  
## Identification Strategy and Estimation Framework

---

## Data Structure

- Panel data: 10,000 users × 35 days
- Observational campaign (non-random exposure)
- Exposure probability correlated with prior engagement and spend

Outcome: Margin dollars  
Treatment: Exposure during promo window

---

## Estimation Strategy

We estimated incremental impact using progressively stronger identification methods.

---

### 1. Naïve Post Comparison

Estimated difference in mean margin between exposed and unexposed users during the promo window.

Limitation:
- Biased due to baseline differences in engagement and spend.

---

### 2. Difference-in-Differences (DiD)

Model:
$$
Y_it = β (Treated_i × Post_t) + α_i + ε_it
$$
Controls for:
- Time-invariant user differences
- Common macro trends

Assumption:
Parallel trends between treated and control groups.

---

### 3. Two-Way Fixed Effects (TWFE)

Model:
$$
Y_it = β (Treated_i × Post_t) + α_i + γ_t + ε_it
$$
Controls for:
- User fixed effects (α_i)
- Date fixed effects (γ_t)

Clustered standard errors at the user level.

---

### 4. Propensity Score Weighting (IPW)

Estimated treatment probability using pre-period covariates:
- Prior 30-day spend
- Engagement score
- Tenure
- Pre-period margin

Computed stabilized inverse probability weights.

Finding:
- Propensity scores clustered near 1.0
- Severe overlap failure
- Covariate balance not materially improved

Conclusion:
IPW not credible due to positivity violation.

---

### 5. Weighted DiD

Applied IPW weights within DiD framework.

Estimate nearly identical to unweighted DiD.

Indicates:
Observable selection not driving remaining bias.

---

### 6. Event Study (Dynamic DiD)

Estimated event-time treatment effects:
$$
Y_it = Σ_k β_k (1[event_time = k] × Treated_i) + α_i + γ_t + ε_it
$$
Findings:
- Pre-period coefficients statistically insignificant.
- No strong evidence of parallel trend violation.
- Immediate and sustained negative margin effect post-promo.

---

## Identification Assumptions

1. Parallel trends (supported by event study).
2. No time-varying unobserved confounders.
3. Stable unit treatment value (no spillovers).
4. Positivity (violated for IPW).

---

## Comparison to Synthetic Ground Truth

True ATT: -6.54  
Estimated DiD: -4.60  

Residual bias likely driven by:
- Treatment effect heterogeneity
- Time-varying targeting intensity
- Structural model assumptions in data-generating process

---

## Conclusion

Difference-in-Differences with fixed effects provides the most credible estimate under observed conditions.

IPW is not reliable due to lack of overlap.

Event study diagnostics support causal interpretation of DiD results.
